```python
# ------------------------VOLUME -------------------------

def HasDuplicate(word_list):
    tuple_list = []
    for sentence in word_list:
        tup = tuple(sentence)
        tuple_list.append(tup)

    c = Counter(tuple_list)
    for key in c.keys():
        if c[key] >= 3:
            return True
    return False

def getMinVolEllipse(P=None, tolerance=0.01):
    """ Find the minimum volume ellipsoid which holds all the points

    Based on work by Nima Moshtagh
    http://www.mathworks.com/matlabcentral/fileexchange/9542
    and also by looking at:
    http://cctbx.sourceforge.net/current/python/scitbx.math.minimum_covering_ellipsoid.html
    Which is based on the first reference anyway!

    Here, P is a numpy array of N dimensional points like this:
    P = [[x,y,z,...], <-- one point per line
         [x,y,z,...],
         [x,y,z,...]]

    Returns:
    (center, radii, rotation)

    """
    global counter2
    (N, d) = np.shape(P)
    d = float(d)

    # Q will be our working array
    Q = np.vstack([np.copy(P.T), np.ones(N)])
    QT = Q.T
#     print(Q.shape)
    # initializations
#     err = 1.0 + tolerance
    err = 1.0
    u = (1.0 / N) * np.ones(N)
#     print(np.diag(u))
    # Khachiyan Algorithm
    while err > tolerance:
        V = Q.dot(np.diag(u)).dot(QT)
#         V = np.dot(Q, np.dot(np.diag(u), QT))
#         print(linalg.det(V))
        try:
            # error here?
            M = np.diag(np.dot(QT , np.dot(linalg.pinv(V), Q)))    # M the diagonal vector of an NxN matrix
        except np.linalg.LinAlgError as err:
            if 'Singular matrix' in str(err):
                print("Singular matrix")
                print(V.shape)
                print(V)
            elif 'SVD did not converge' in str(err):
                print("SVD did not converge")
                print(V.shape)
                print(V)
                print("Error at record ", counter2)
                return (0,0,0)

            else:
                raise

        j = np.argmax(M)
        maximum = M[j]
        step_size = (maximum - d - 1.0) / ((d + 1.0) * (maximum - 1.0))
        new_u = (1.0 - step_size) * u
        new_u[j] += step_size
        err = np.linalg.norm(new_u - u)
#         print(err)
        u = new_u

    # center of the ellipse
    center = np.dot(P.T, u)

    # the A matrix for the ellipse
    try:
        A = linalg.pinv(
                    np.dot(P.T, np.dot(np.diag(u), P)) -
                    np.array([[a * b for b in center] for a in center])
                    ) / d
    except np.linalg.LinAlgError as err:
        if 'SVD did not converge' in str(err):
            print("SVD did not converge")
            print("Error at record ", counter2)
            return (0,0,0)

    # Get the values we'd like to return
    U, s, rotation = linalg.svd(A)
    radii = 1.0/np.sqrt(s)
#     print(s.shape)
    return (center, radii, rotation)

def getEllipsoidVolume(radii, embedding_dimension, N):
    """Calculate the volume of the blob"""
    dimension = N - 1
    if dimension == 0 or dimension == 1:
        #return -np.inf
        return 0

    if dimension > embedding_dimension:
        dimension = embedding_dimension

    coeff = (2*math.pow(np.pi,dimension/2))/(dimension*math.gamma(dimension/2))
#     print(coeff)
    log_coeff = np.log(coeff)
    effectice_radii = radii[-dimension:]
    log_rad = np.log(effectice_radii)
    logsum = np.sum(log_rad)
    logv = log_coeff + logsum
#     print()
#     print(logv)
#         v = math.exp(logv)
#         print(v)
    return logv

def calc_vol_speed_circ(word_list, tag_dict=tag_dict, embedding_matrix=english_embeddings, embedding_idx=index, tolerance=0.01):
    #word_list is a list of lists
    global counter1
    counter1+=1
    avg_sent_vecs=[]
    speed_list=[]
    print("vol_original row number",counter1)
    Has_Duplicate = HasDuplicate(word_list)
    if Has_Duplicate:
        print('-------------------------------')
        print('duplicate detected!')
        print('-------------------------------')

    for sentence in word_list:
        embedding_vectors = []
        for word_tag_tuple in sentence:
            if word_tag_tuple[1] in tag_dict:
                word = word_tag_tuple[0]
                if word in embedding_idx:
                    embedding_vectors.append(embedding_matrix[embedding_idx[word]])

        embedding_vectors = np.asarray(embedding_vectors)
        # note nothing happens if shape = 0
        temp_list=[]
        if embedding_vectors.shape[0] >= 1:
            temp_array = np.array(embedding_vectors).mean(axis=0)
            temp_list = temp_array.tolist()
            avg_sent_vecs.append(temp_list)
    avg_sent_vecs = np.asarray(avg_sent_vecs)

    #speed and circuitousness calculations
    n = 0
    circ = 0
    for row in avg_sent_vecs:
        if n==0:
            old_row = row
            n+=1
        else:
            new_row = row
            distance = scipy.spatial.distance.euclidean(old_row,new_row)
            speed_list.append(distance)
            old_row = row
            n+=1
    if len(speed_list) > 0:
        if Has_Duplicate:
            avg_speed = 0
            circ = -1
            route_dist_div = -1
        else:
            avg_speed = sum(speed_list)/len(speed_list)

            #TSP solution: see https://stackoverflow.com/questions/71196078/scalable-implementation-of-tsp-in-python
            # create distance matrix
            distances = {}
            u = 0
            last_index = len(avg_sent_vecs)-1
            for row in avg_sent_vecs:
                distances[u] = {}
                v = 0
                while(v <= last_index):
                    if u == v:
                        distances[u][v] = 0
                    else:
                        #multiply by 1 million because the distance matrix has to be integers
                        d = np.floor(scipy.spatial.distance.euclidean(avg_sent_vecs[u],avg_sent_vecs[v])*1000000).astype(int)
                        distances[u][v] = d
                    v+=1
                u+=1

            manager = pywrapcp.RoutingIndexManager(len(avg_sent_vecs), 1, [0], [last_index]) #num cities, vehicles, depot

            # Create Routing Model.
            routing = pywrapcp.RoutingModel(manager)

            def distance_callback(from_index, to_index):
                    """Returns the distance between the two nodes."""
                    # Convert from routing variable Index to distance matrix NodeIndex.
                    from_node = manager.IndexToNode(from_index)
                    to_node = manager.IndexToNode(to_index)
                    return distances[from_node][to_node]

            transit_callback_index = routing.RegisterTransitCallback(distance_callback)

            # Define cost of each arc.
            routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

            # Setting first solution heuristic.
            search_parameters = pywrapcp.DefaultRoutingSearchParameters()
            search_parameters.first_solution_strategy = (
                routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)

            # Solve the problem.
            solution = routing.SolveWithParameters(search_parameters)

        # Print solution on console.
            if solution:
                #print('Objective: {}'.format(solution.ObjectiveValue()))
                index = routing.Start(0)
                plan_output = 'Route:\n'
                route_distance = 0
                while not routing.IsEnd(index):
                    plan_output += ' {} ->'.format(manager.IndexToNode(index))
                    previous_index = index
                    index = solution.Value(routing.NextVar(index))
                    route_distance += routing.GetArcCostForVehicle(previous_index, index, 0)
                plan_output += ' {}\n'.format(manager.IndexToNode(index))
                #print(plan_output)
                plan_output += 'Objective: {}m\n'.format(route_distance)

                route_dist_div = route_distance/1000000
                if route_distance == 0:
                    circ = -1
                else:
                    circ = sum(speed_list)/(route_dist_div)
    else:
        avg_speed = 0
        circ = -1
        route_dist_div = -1

    # volume calculations
    avg_vec_volume = 0
    if avg_sent_vecs.shape[0] <= 1 or Has_Duplicate:
        avg_vec_volume = 0
    else:
        center, radii, rotation = getMinVolEllipse(avg_sent_vecs, tolerance)
        if isinstance(radii,int):
            if radii == 0:
                #return 0
                avg_vec_volume = 0
            else:
                avg_vec_volume = getEllipsoidVolume(radii, embedding_matrix.shape[1], len(avg_sent_vecs))
    return pd.Series([avg_vec_volume, avg_speed, circ, route_dist_div])
```
